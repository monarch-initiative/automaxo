```python
maxo_annotations_df = pd.read_csv('maxo-annotations.tsv', delimiter='\t')
evaluation_no_replacement_df = pd.read_csv('evaluation_no_replacement.tsv', delimiter='\t')
```

```python
maxo_annotations_desired_df = maxo_annotations_df[['disease_id', 'disease_name', 'citation', 'maxo_id', 'maxo_label', 'hpo_id', 'maxo_relation']]
```

```python
annotated_pm_ids = maxo_annotations_desired_df['citation']
annotated_pm_ids_list = annotated_pm_ids.tolist()
# Extract numeric part, remove duplicates, and filter out invalid elements
annotated_pm_ids = set(id.split(':')[1] for id in annotated_pm_ids_list if id.startswith('PMID:'))
len(annotated_pm_ids)
```

```python
retrieved_pmids= set(evaluation_no_replacement_df.iloc[:, 0])
# Convert set1 to a set of strings
retrieved_pmids = {str(item) for item in retrieved_pmids}

len(retrieved_pmids)
```

```python
non_retrieved_pmids =  annotated_pm_ids - retrieved_pmids
```

```python
import pandas as pd
import json
```

```python
# Read final Json File
post_ontoGPT_file_path = "evaluation_post_ontoGPT.json"
```

```python
# Load JSON data from file
with open(post_ontoGPT_file_path, 'r') as file:
    data = json.load(file)
```

```python
# Initialize an empty list to store rows
rows = []

# Iterate through the JSON data and extract the relevant information
for entry in data.get('ranked_triplets'):
    triplet = entry.get('triplet', [])
    if len(triplet) == 3:
        subject_with_label, predicate, object_with_label = triplet
        subject, subject_label = subject_with_label.split(': ', 1)
        object_, object_label = object_with_label.split(': ', 1)
        for pubmed_id in entry.get('pubmed_ids_mesh_info', {}):
            rows.append([subject, subject_label, predicate, object_, object_label, pubmed_id])

# Create a DataFrame from the rows
df = pd.DataFrame(rows, columns=['Subject', 'Subject Label', 'Predicate', 'Object', 'Object Label', 'Citation'])



# Rename 'Subject Label' to 'maxo_label'
df.rename(columns={'Subject Label': 'maxo_label'}, inplace=True)

# Initialize new columns
df['maxo_id'] = ''
df['disease_id'] = ''
df['hpo_id'] = ''
df['disease_name'] = ''
df['hpo_label'] = ''
df['mondo_relationship'] = ''
df['hpo_relationship'] = ''

# Function to extract terms and set relationship columns
def extract_terms(row):
    if 'MAXO:' in row['Subject']:
        row['maxo_id'] = row['Subject']
        row['Subject'] = ''
    if 'MONDO:' in row['Object']:
        row['disease_id'] = row['Object']
        row['disease_name'] = row['Object Label']
        row['mondo_relationship'] = row['Predicate']
        row['Object'] = ''
    if 'HP:' in row['Object']:
        row['hpo_id'] = row['Object']
        row['hpo_label'] = row['Object Label']
        row['hpo_relationship'] = row['Predicate']
        row['Object'] = ''
    return row

# Apply function to each row
df_new = df.apply(extract_terms, axis=1)

# Reorder columns
df_new = df_new[['Citation', 'maxo_id', 'maxo_label', 'Subject', 'Predicate', 'disease_id', 'disease_name', 'mondo_relationship', 'hpo_id', 'hpo_label', 'hpo_relationship', 'Object']]
```

```python
df_new
```

```python
# Group by 'Citation', 'maxo_id', and 'maxo_label'
grouped_df = df_new.groupby(['Citation', 'maxo_id', 'maxo_label'])

# Aggregate other columns
merged_df = grouped_df.agg({
    'Subject': 'first',  # Keep the first non-null value
    'Predicate': 'first',  # Keep the first non-null value
    'disease_id': lambda x: ', '.join(x.dropna()),  # Join non-null values with comma
    'disease_name': lambda x: ', '.join(x.dropna()),  # Join non-null values with comma
    'mondo_relationship': lambda x: ', '.join(x.dropna()),  # Join non-null values with comma
    'hpo_id': lambda x: ', '.join(x.dropna()),  # Join non-null values with comma
    'hpo_label': lambda x: ', '.join(x.dropna()),  # Join non-null values with comma
    'hpo_relationship': lambda x: ', '.join(x.dropna()),  # Join non-null values with comma
    'Object': 'first'  # Keep the first non-null value
}).reset_index()

```

```python
merged_df
```

```python
# Save the post-OntoGPT triplet DataFrame to a TSV file
merged_df.to_csv('results/post_ontoGPT_merged.tsv', sep='\t', index=False)
```
